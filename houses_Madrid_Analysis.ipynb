**Allotment Area**
An allotment area (or agricultural plot) is a piece of land that is either privately leased from an individual owner or rented from local authorities for the purpose of cultivating fruits, vegetables, and other crops. In some cases, this land is also used for ornamental plants, as well as raising chickens, rabbits, and bees. Traditionally, the agricultural area is measured in square feet (sq ft) or acres, with 10 feet being a commonly accepted measurement for an allotment plot, which is approximately equivalent to 250 square meters or the size of a double tennis court.

## *Apply this step* 
* ### remove Nan Columns
* ### Get Specific columns for processing
* ### Check if there are duplicated row

```python
# axis = 1 mean remove columns but axis = 0 mean remove row
# and how = all mean remove this column if all value = NULL but if how =any this mean remove this column if just found at least one value in this column = NULL
 
data = data.dropna(axis=1,how='all')
data.info()
```

```python
data.duplicated().sum()
```

```python
# Taking only needed columns from data
df = data[['id','subtitle','sq_mt_built','sq_mt_useful','n_rooms','n_bathrooms','n_floors','sq_mt_allotment','rent_price','buy_price','buy_price_by_area','built_year','parking_price']]
df.info()
```

* ### Change `subtitle` column into `city, raw_address` into `address`
* ### Check what you will do in NaN Values

```python
df = df.rename(columns={'subtitle': 'city'})
```

```python
# Remove Columns that have a lot of NaN Values 
# (sq_mt_useful , n_floors , sq_mt_allotment , built_year , parking_price)
df = df[['id','city','sq_mt_built','n_rooms','n_bathrooms','rent_price','buy_price','buy_price_by_area']]
df.info()
```

```python
df.n_rooms.describe()
```

### we found `n_rooms = 0` and this not logic we need remove this row contain 0 room

```python
# Remove Rows with 0 rooms
df.drop(df.index[df.n_rooms == 0],axis=0,inplace=True)
```

```python
# Add Mode into NaN cells in n_bathrooms column
df.n_bathrooms.mode()
```

```python
# fill Nan value by Mode = 1 
# and replace value from 0 to 1 because we need min value is 1 bathroome
df.n_bathrooms = df.n_bathrooms.fillna(1)
df.n_bathrooms = df.n_bathrooms.replace(0,1)
```

```python
# Convert n_bathrooms type into integer
df.n_bathrooms = df.n_bathrooms.astype(int)
```

```python
df.info()
```

```python
# Drop sq_mt_built NaN rows because this column is very effective on result so can't complete this column by any value not accurate
df.dropna(subset='sq_mt_built',axis=0,inplace=True)
```

```python
df.info()
```

```python
df.sample(10)
```

```python
# Change negative values in rent price column to Zero FOR NOW

for x in df.rent_price.values:
    if x < 0:
        i = df.index[df.rent_price == x]
        df.rent_price[i] = 0
```

# ***Linear Regression Model***

```python
reco = df.query('district == "Recoletos"')
```

```python
reco.sample(10)
```

```python
reco.info()
```

```python
import statsmodels.api as sm

reco["intercept"]=1
lm = sm.OLS(reco['buy_price'],reco[['intercept','sq_mt_built']])
result = lm.fit()
result.summary()
```

### intercept Coef this mean --> base (b)
### sq_mt_built Coef this mean --> weight (w)
### and b and w is a parameter used in linear regression equation --> ( w * x + b)

```python
# example here b = -43420 and w = 8975.6204 take this value from above summary
x= 300
b = -43420
w=8975.6204
price = w*x+b
print('Predicted Price of Flat in Recoletos with are of {} m2 = {} Euro'.format(x,price))
```
